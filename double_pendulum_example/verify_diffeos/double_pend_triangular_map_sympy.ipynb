```python
%load_ext autoreload
%autoreload 2
import numpy as np
import sympy as sp
q1, q2, p1, p2 = sp.symbols('q1 q2 p1 p2', real=True)
variables = [q1, q2, p1, p2]
H_sym = (p1**2 + 2*p2**2 - 2*sp.cos(q1 - q2)*p1*p2)/(4 - 2*sp.cos(q1 - q2)**2)  + 1/2*q1**2 + 1/2*q2**2
invariants = [H_sym]
H_fun = sp.lambdify(variables, H_sym, 'numpy')
```

```python
q1, q2, p1, p2, t, k = sp.symbols('q1 q2 p1 p2 t k', real=True)
variables = [q1, q2, p1, p2]

def G(y):
    y1,y2,y3,y4 = y
    return y3**2 + y4**2 + 1/2*y1**2 + 1/2*y2**2

def phi(x):
    x1,x2,x3,x4 = x
    c12 = sp.cos(x1 - x2)
    D   = sp.sqrt(2.0 - c12*c12)
    y1  = x1
    y2  = x2
    y3  =  0.5*x3
    y4  = (2.0*x4 - c12*x3)/(2.0*D)
    return sp.Matrix([y1,y2,y3,y4])

def phi_inv(y, x_prev=None):
    y1,y2,y3,y4 = y
    x1 = y1
    x2 = y2
    x3 = 2.0*y3
    c12 = sp.cos(x1 - x2)
    D   = sp.sqrt(2.0 - c12*c12)
    x4  = 0.5*c12*x3 + D*y4
    return sp.Matrix([x1,x2,x3,x4])

phi_inv(phi(variables))
```

```python

# H_expr
exp = sp.exp(t)

H = lambda x: H_sym.subs({q1: x[0], q2: x[1], p1: x[2], p2: x[3]})
print('conjugate invariant: ', sp.simplify(G(phi(variables))-H(variables)))
print('homogeneous in conjugate space: ', sp.expand(G(exp*phi(variables)) - sp.exp(k*t)*G(phi(variables))))
print('c-homogeneous in physical space: ', sp.factor(H(phi_inv(exp*phi(variables)))-sp.exp(k*t)*H(variables)))
```
